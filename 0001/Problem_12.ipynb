## Highly divisible triangular number



The sequence of triangle numbers is generated by adding the natural numbers. So the 7th triangle number would be 1 + 2 + 3 + 4 + 5 + 6 + 7 = 28. The first ten terms would be:

1, 3, 6, 10, 15, 21, 28, 36, 45, 55, ...

Let us list the factors of the first seven triangle numbers:

     1: 1
     3: 1,3
     6: 1,2,3,6
    10: 1,2,5,10
    15: 1,3,5,15
    21: 1,3,7,21
    28: 1,2,4,7,14,28

We can see that 28 is the first triangle number to have over five divisors.

What is the value of the first triangle number to have over five hundred divisors?




```python
%%time

import math
import numpy as np

i = 1
triangle_num1 = 1
solution_found = False

def get_primefactors(number):
    # Find primes with sieve of eratosthenes
    primes = np.zeros(np.int(number) + 1)
    primes[0] = -1
    primes[1] = -1
    for i in range(2, np.int(math.sqrt(number))):
        newprime = i
        notprime = i*i
        while notprime <= number:
            primes[notprime] = -1
            notprime = notprime + newprime
    return primes
    
primes = get_primefactors(1000)

while solution_found == False:
    # Generate next triangle number
    i = i+1
    triangle_num1 = triangle_num1 + i
    triangle_num = triangle_num1
    prime_count = np.copy(primes)
    #print("Next triangle_number {}".format(triangle_num))
    
    # Test for number of divisors
    # Find prime factorization
    j = 2
    while j < primes.size:
        if prime_count[j] >= 0 and triangle_num%j == 0:
            #print("Testing prime number {}".format(j))
            prime_count[j] = prime_count[j] + 1
            #print("Prime count is {}".format(prime_count[j]))
            triangle_num = triangle_num/j
        else:
            j = j + 1
                
            
    # Test for number of divisors
    num_divisors = 1
    for j in range(primes.size):
        if prime_count[j] > 0:
            num_divisors = num_divisors*(prime_count[j] + 1)
            
    if num_divisors >= 500:
        solution_found = True
        print(triangle_num1)
        print("Number of divisors {}".format(num_divisors))
    
```
